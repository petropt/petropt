# Your First Petroleum Dataset in Python

This notebook introduces `petropt`, the Python package for petroleum engineering.
You'll learn how to:
1. Load production data
2. Calculate PVT properties
3. Build an IPR curve
4. Visualize results

```python
import petropt
import matplotlib.pyplot as plt
import numpy as np

print(f"petropt version: {petropt.__version__}")
```

## 1. Load Sample Production Data

`petropt` comes with bundled sample datasets for learning.

```python
df = petropt.datasets.load_sample_production()
df.head(10)
```

```python
# Plot production for Well-A
well_a = df[df['well_name'] == 'Well-A']

fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.plot(well_a['date'], well_a['oil'], 'g-o', label='Oil (BOPD)')
ax1.plot(well_a['date'], well_a['water'], 'b-s', label='Water (BWPD)')
ax1.set_xlabel('Date')
ax1.set_ylabel('Rate (bbl/day)')
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(well_a['date'], well_a['gas'], 'r-^', label='Gas (MCFD)')
ax2.set_ylabel('Gas Rate (MCFD)')
ax2.legend(loc='upper right')

plt.title('Well-A Production History')
plt.tight_layout()
plt.show()
```

## 2. PVT Correlations

Calculate fluid properties using Standing's correlations.

```python
# Reservoir fluid properties
api = 35          # API gravity
gas_sg = 0.65     # Gas specific gravity
temp = 200        # Reservoir temperature (°F)

# Bubble point pressure
pb = petropt.correlations.standing_bubble_point(api=api, gas_sg=gas_sg, temp=temp)
print(f"Bubble point pressure: {pb:.0f} psi")

# Solution GOR at bubble point
rs = petropt.correlations.standing_rs(pressure=pb, temp=temp, api=api, gas_sg=gas_sg)
print(f"Solution GOR at Pb: {rs:.0f} scf/STB")

# Oil FVF at bubble point
bo = petropt.correlations.standing_bo(rs=rs, temp=temp, api=api, gas_sg=gas_sg)
print(f"Oil FVF at Pb: {bo:.4f} bbl/STB")

# Oil viscosity
visc = petropt.correlations.beggs_robinson_viscosity(temp=temp, api=api, rs=rs)
print(f"Dead oil viscosity: {visc['dead_oil_viscosity_cp']:.2f} cp")
print(f"Live oil viscosity: {visc['live_oil_viscosity_cp']:.2f} cp")
```

## 3. Inflow Performance Relationship (IPR)

Build a Vogel IPR curve for a solution-gas-drive well.

```python
# Generate IPR curve
qmax = 1500  # AOF in STB/day
pr = 3500    # Reservoir pressure in psi

ipr = petropt.correlations.vogel_ipr(qmax=qmax, pr=pr, num_points=50)

plt.figure(figsize=(8, 6))
plt.plot(ipr['qo'], ipr['pwf'], 'b-', linewidth=2)
plt.xlabel('Oil Rate (STB/day)')
plt.ylabel('Flowing Bottomhole Pressure (psi)')
plt.title('Vogel IPR Curve')
plt.grid(True, alpha=0.3)
plt.xlim(0)
plt.ylim(0)
plt.tight_layout()
plt.show()

print(f"AOF (qmax): {qmax} STB/day")
print(f"Reservoir pressure: {pr} psi")
```
